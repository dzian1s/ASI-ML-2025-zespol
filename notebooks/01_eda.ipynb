```python
data = data.dropna()
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

num_cols = data.select_dtypes(include=['int64', 'float64']).columns
cat_cols = data[['neighbourhood', 'neighbourhood_group', 'room_type']]

# Wykresy dla cech liczbowych
for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(data[col], kde=True, bins=30, color='steelblue')
    plt.title(f"Rozkład cechy: {col}")
    plt.xlabel(col)
    plt.ylabel("Liczba obserwacji")
    plt.tight_layout()
    plt.show()

# Wykresy dla cech kategorycznych
for col in cat_cols:
    plt.figure(figsize=(7, 4))
    order = data[col].value_counts().index
    sns.countplot(data=data, x=col, order=order, palette='muted')
    plt.title(f"Rozkład cechy kategorycznej: {col}")
    plt.xlabel(col)
    plt.ylabel("Liczba ofert")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
usunięcie zbędnych zmiennych - id, name, host_id, host_name, last_review - id -> jest unikalne, name -> również unikalny i za długi do enkodowania, last_review -> duzo unikalnych
```

```python
data_dropped = data.drop(columns=['id', 'name', 'host_id', 'host_name', 'last_review'])
```

```python
data_dropped
```

enkodowanie label encoding - będziemy używać RandomForrestRegressor

```python
from sklearn.preprocessing import LabelEncoder

le_neigh = LabelEncoder()
data_dropped['neighbourhood'] = le_neigh.fit_transform(data_dropped['neighbourhood'])

le_neigh_group = LabelEncoder()
data_dropped['neighbourhood_group'] = le_neigh_group.fit_transform(data_dropped['neighbourhood_group'])

le_room = LabelEncoder()
data_dropped['room_type'] = le_room.fit_transform(data_dropped['room_type'])
```

```python
data_dropped
```

wybieramy MAE(Mean Absolute Error) i RMSE(Root Mean Square Error) -> bedziemy przewidywać cenę więc najlepsze metryki do regresji

```python
data_dropped.to_csv('../data/02_interim/airbnb_cleaned.csv', index=False)
```
